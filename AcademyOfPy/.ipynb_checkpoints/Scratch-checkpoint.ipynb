```python
import pandas as pd
import numpy as np

file_school_data = "Resources/schools_complete.csv"
file_student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(file_school_data)
student_data = pd.read_csv(file_student_data)

# Combine the data into a single dataset
#combined_df = pd.merge(student_data, school_data, how="left", on="school_name")
combined_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
ninth_grade = combined_df[combined_df.grade == '9th'][["school_name","math_score"]].set_index("school_name").groupby("school_name").mean()
ninth_grade.columns = ['9th']

tenth_grade = combined_df[combined_df.grade == '10th'][["school_name","math_score"]].set_index("school_name").groupby("school_name").mean()
tenth_grade.columns = ['10th']

eleventh_grade = combined_df[combined_df.grade == '11th'][["school_name","math_score"]].set_index("school_name").groupby("school_name").mean()
eleventh_grade.columns = ['11th']

twelfth_grade = combined_df[combined_df.grade == '12th'][["school_name","math_score"]].set_index("school_name").groupby("school_name").mean()
twelfth_grade.columns = ["12th"]
# twelfth_grade.columns

combinded_grades = pd.merge(ninth_grade, tenth_grade, on="school_name")
combinded_grades = pd.merge(combinded_grades, eleventh_grade, on="school_name")
combinded_grades = pd.merge(combinded_grades, twelfth_grade, on="school_name")
combinded_grades.index.names = ['School Name']

combinded_grades["9th"] = combinded_grades["9th"].map("{:.2f}".format)
combinded_grades["10th"] = combinded_grades["10th"].map("{:.2f}".format)
combinded_grades["11th"] = combinded_grades["11th"].map("{:.2f}".format)
combinded_grades["12th"] = combinded_grades["12th"].map("{:.2f}".format)

combinded_grades.index.names = ['School Name']

combinded_grades
```
